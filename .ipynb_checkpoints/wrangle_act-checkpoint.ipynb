** <h3> TWEET_ARC - LIMPANDO **

**Problema 1 - Mudando tweet_id, in_reply e retweet para string **

```python
twitter_arc_clean['tweet_id '] = twitter_arc_clean.tweet_id.astype(str)
```

```python
twitter_arc_clean["in_reply_to_status_id"] = twitter_arc_clean["in_reply_to_status_id"].astype(str)
twitter_arc_clean["retweeted_status_id"] = twitter_arc_clean["retweeted_status_id"].astype(str)
```

```python
Replies = twitter_arc_clean[twitter_arc_clean["in_reply_to_status_id"] != "nan"]
retw = twitter_arc_clean[twitter_arc_clean["retweeted_status_id"] != "nan"]
```

```python
twitter_arc_clean.drop(Replies["tweet_id"].index,inplace=True)
twitter_arc_clean.drop(retw["tweet_id"].index,inplace=True)b
```

** Problema 2 - Idenficando o nome dos animais e mudando na tabela name **

```python
Name = twitter_arc_clean["text"].str.contains("This is ([A-Z][a-z])+")
```

```python
twitter_arc_clean = twitter_arc_clean[Name]
```

```python
twitter_arc_clean["name"] = twitter_arc_clean.text.str.extract("This is ([A-Z][a-z]+).", expand=False)
```

** Problema 3 - Resumindo os atributos doggo,floofer,pupper,puppo em um só atributo **

```python
state_doggo = twitter_arc_clean[twitter_arc_clean["doggo"] != "None" ]["tweet_id"].index
state_floofer = twitter_arc_clean[twitter_arc_clean["floofer"] != "None" ]["tweet_id"].index
state_pupper = twitter_arc_clean[twitter_arc_clean["pupper"] != "None" ]["tweet_id"].index
state_puppo = twitter_arc_clean[twitter_arc_clean["puppo"] != "None" ]["tweet_id"].index
```

```python
Dog_state = []
for i in range(len(twitter_arc_clean)):
    if i in state_doggo:
            Dog_state.append("Doggo")
    else:
        if i in state_floofer:
            Dog_state.append("Floofer")
        else:
            if i in state_pupper:
                    Dog_state.append("Pupper")
            else:
                if i in state_puppo:
                    Dog_state.append("Puppo")
                else:
                    Dog_state.append("None")
```

```python
twitter_arc_clean["Dog_state"] = pd.Series(Dog_state)
```

** Problema 4 - Removendo os atributos desnecessários **

```python
Remover_col = ["timestamp","expanded_urls", "doggo", "floofer", "pupper", "puppo",
                   "in_reply_to_status_id", "in_reply_to_user_id", 
                   "retweeted_status_id", "retweeted_status_user_id", 
                    "retweeted_status_timestamp"]
twitter_arc_clean.drop(columns=Remover_col,inplace=True)
```

**Problema 5 - Retirando informações desnecessários no texto e no source **

```python
len_html = len(twitter_arc_clean["text"][0].split()[-1])+1
twitter_arc_clean["text"] = twitter_arc_clean["text"].apply(lambda x: x[:-len_html])
```

```python
source = twitter_arc_clean.source.values
```

```python
for i in range(len(source)):
    source[i] = source[i].split('"')[1]
```

** Problema 6 - Escolhando somente os twitters com denominados igual a 10 **

```python
Denominador_correto = twitter_arc_clean.text.str.contains("\d+/10")
```

```python
twitter_arc_clean = twitter_arc_clean[Denominador_correto]
```

** Problema 7 - Removendo coluna denominator e mudando a coluna tweet_id para string **

```python
twitter_arc_clean["tweet_id"] = twitter_arc_clean.tweet_id.astype(str)
```

```python
twitter_arc_clean.drop(columns=['rating_denominator'],inplace=True)
```
